```python
import pandas as pd

dtype_map = {
    'order_item_id': 'string',
    'order_id': 'string',
    'product_id': 'string',
    'quantity': 'string',
    'unit_price': 'string',
    'line_total': 'string',
    'discount_amount': 'string'
}

df_order_items = pd.read_csv('../../data/raw/order_items.csv', dtype=dtype_map)
df_order_items.head()
```

Vamos a descartar las columnas que son innecesarias para nuestro an√°lisis.

```python
df_order_items.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df_order_items_clean = df_order_items.set_index('order_item_id')

# INTEGERS
for col in ['order_id', 'product_id', 'quantity']:
    df_order_items_clean[col] = df_order_items_clean[col].str.strip()
    df_order_items_clean[col] = pd.to_numeric(df_order_items_clean[col], errors='coerce').astype('Int64')

# FLOATS
for col in ['unit_price', 'line_total', 'discount_amount']:
    df_order_items_clean[col] = df_order_items_clean[col].str.strip()
    df_order_items_clean[col] = pd.to_numeric(df_order_items_clean[col], errors='coerce')

df_order_items_clean.head()
```

Exportamos el dataset limpio para reutilizarlo.

```python
df_order_items_clean.to_pickle('../../data/clean/order_items.pkl')
```
